```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import csv
```

```python
driver=webdriver.Chrome()
```

```python
# Click on View more Stories Button

driver.get(url="https://www.cnbctv18.com/economy/")
wait = WebDriverWait(driver, 10)  
```

```python
articles_path=f'/html/body/div[1]/div[3]/div[2]/div/div[7]/section/article[1]/div/a'           
print(articles_path)
articles = driver.find_element(By.XPATH, articles_path) 
print(articles.text)
```

```python
articles_dict = {}
k = 0
iter_flag = True
wait=WebDriverWait(driver,10)

for i in range(7,31,2):
        for j in range(1,15): 
            articles_path=f'/html/body/div[1]/div[3]/div[2]/div/div[{i}]/section/article[{j}]/div/a'
            print(articles_path)
            articles = wait.until(EC.visibility_of_element_located((By.XPATH,articles_path)))
            publish_details=driver.find_element(By.XPATH,f'/html/body/div[1]/div[3]/div[2]/div/div[{i}]/section/article[{j}]/div/a/div/span[1]/time')
    
            details = {
                'id': k,
                'publisher': "CNBC",
                'title': articles.get_attribute('title'),
                'link': articles.get_attribute('href'),
                'publish_details': publish_details.text,
                'content': ""
            }
            
            articles_dict[f'{k}'] = details
            k=k+1
        driver.implicitly_wait(8)     

print(articles_dict)
```

```python
from newspaper import Article

def extract_content(url):
    # Create an Article object
    article = Article(url)

    # Download and parse the article
    article.download()
    article.parse()
    
    return article.text

for keys in articles_dict.keys():
    link=articles_dict[keys]['link']
    # driver.get(url=link)    
    # date=driver.find_element(By.XPATH,"/html/body/section[1]/section/div[3]/div/div[2]/div[1]").text
    # content= driver.find_element(By.XPATH,'/html/body/section[1]/section/div[3]/div').text
    content=extract_content(link)
    
    articles_dict[keys]['content']= content
```

```python
print(articles_dict)
```

```python
# Flatten the nested structure
flat_articles_list = []
for article_id, article_data in articles_dict.items():
    flat_article = {'id': article_data['id'], 'publisher': article_data['publisher'], 'title': article_data['title'], 'link': article_data['link'], 'publish_details': article_data['publish_details'], 'content': article_data['content']}
    flat_articles_list.append(flat_article)

# Specify the CSV file path
csv_file_path = 'cnbc.csv'

# Write the flattened data to a CSV file
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    fieldnames = ['id', 'publisher', 'title', 'link', 'publish_details', 'content']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    
    # Write the header
    writer.writeheader()
    
    # Write the data
    writer.writerows(flat_articles_list)

print(f"The data has been written to '{csv_file_path}'.")
```
